```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

```python
#Reading Whale Returns
whale_returns_csv = Path(r"C:\Users\aarora\OneDrive - AMERIBEST HOMECARE\Desktop\New folder\whale_returns.csv")
whale_returns = pd.read_csv(whale_returns_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
whale_returns.sort_index(ascending = True, inplace = True)
whale_returns.head()
```

```python
# count null
print(whale_returns.isnull().sum())
```

```python
# drop null
whale_returns.dropna(inplace=True)
whale_returns.isnull().sum()
```

```python
#reading algo returns
algo_returns_csv = Path(r"C:\Users\aarora\OneDrive - AMERIBEST HOMECARE\Desktop\New folder\algo_returns.csv")
algo_returns = pd.read_csv(algo_returns_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
algo_returns.sort_index(ascending = True, inplace = True)
algo_returns.head()
```

```python
# count null
algo_returns.isnull().sum()
```

```python
# drop null
algo_returns.dropna(inplace = True)
algo_returns.head()
```

```python
#reading SP500 returns
sp500_history_csv = Path(r"C:\Users\aarora\OneDrive - AMERIBEST HOMECARE\Desktop\New folder\sp500_history.csv")
sp500_history = pd.read_csv(sp500_history_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
sp500_history.sort_index(ascending = True, inplace = True)
sp500_history.head()
```

```python
#Daily Returns
sp500_daily_returns = sp500_history.pct_change()
```

```python
# drop null
sp500_daily_returns.dropna(inplace = True)
sp500_daily_returns.head()
```

```python
#rename close column
sp500_daily_returns.rename(columns = {"Close":"sp500"}, inplace = True)
```

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
df_daily = pd.concat([whale_returns, algo_returns,sp500_daily_returns], axis = "columns",join = "inner")
```

```python
# Plot daily returns
df_daily.plot(figsize = (20,10))
```

```python
# Plot cumulative returns
df_cum = (1+ df_daily).cumprod()
df_cum.plot(figsize = (20,10))
```

```python
# Box plot to visually show risk
df_daily.plot(kind = "box", figsize = (20,10))
```

```python
# Calculate the daily standard deviations of all portfolios
df_daily_std = pd.DataFrame(df_daily.std()).rename(columns = {0:"std"})
df_daily_std
```

```python
# Determine which portfolios are riskier than the S&P 500
higher_std = df_daily_std[df_daily_std["std"] > df_daily_std.loc["sp500", "std"]]
higher_std
```

```python
# Calculate the annualized standard deviation (252 trading days)
df_annualized_std = df_daily_std*np.sqrt(252)
df_annualized_std
```

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
sp500_rolling_21 = df_daily[["sp500"]].rolling(window = 21).std()
sp500_rolling_21.plot(figsize = (20,10))
```

```python
# Calculate the correlation
correlation_sp500 = pd.DataFrame(df_daily.corr()).loc[:,"sp500"]
correlation_sp500[correlation_sp500==correlation_sp500[correlation_sp500<1].max()]
```

```python
df_daily.plot(kind="scatter", y = "Algo 2", x = "sp500", figsize = (20,10))
```

```python
#covariance of a portfolio
rolling_covariance = df_daily['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(df_daily['sp500'])
rolling_variance = df_daily['sp500'].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 60 Day Beta of B.H')
```

```python
#rolling window
df_daily_ewm = df_daily.ewm(span = 21,adjust = False).mean()
(1+df_daily_ewm).cumprod().plot(figsize = (20,10))
```

```python
#sharpe ratios
sharpe_ratios = df_daily.mean()*252/(df_daily.std()*np.sqrt(252))
sharpe_ratios
```

```python
sharpe_ratios.plot(kind = "bar", title = "Sharpe Ratios Result")
```

```python
#my portfolio and stocks
#stock 1
aapl_history = pd.read_csv(Path(r"C:\Users\aarora\OneDrive - AMERIBEST HOMECARE\Desktop\New folder\aapl_history.csv"), index_col="Date", parse_dates = True, infer_datetime_format= True)
aapl_history.rename(columns = {"NOCP":"aapl"}, inplace = True)
aapl_history.head()
```

```python
#stock 2
cost_history = pd.read_csv(Path(r"C:\Users\aarora\OneDrive - AMERIBEST HOMECARE\Desktop\New folder\cost_history.csv"), index_col="Date", parse_dates = True, infer_datetime_format= True)
cost_history.rename(columns = {"NOCP":"cost"}, inplace = True)
cost_history.head()
```

```python
#stock 3
goog_history = pd.read_csv(Path(r"C:\Users\aarora\OneDrive - AMERIBEST HOMECARE\Desktop\New folder\goog_history.csv"), index_col="Date", parse_dates = True, infer_datetime_format= True)
goog_history.rename(columns = {"NOCP":"goog"}, inplace = True)
goog_history.head()
```

```python
#combine 
my_stocks = pd.concat((aapl_history, cost_history, goog_history), axis = 1, join= "inner")
my_stocks.sort_index(ascending = True, inplace = True)
my_stocks.head()
```

```python
# Calculate Daily Returns
my_stocks = my_stocks.pct_change()
# Drop nulls
my_stocks.dropna(inplace = True)
my_stocks.head()
```

```python
my_stocks.plot(figsize = (20,10))
```

```python
#weighted returns
weights = [1/3, 1/3, 1/3]
my_portfolio = my_stocks.dot(weights)
my_portfolio_returns = my_portfolio.pct_change()
my_portfolio_returns.dropna(inplace = True)
my_portfolio_returns.head()
```

```python
#combine dataframe with previous dataframe
portfolio_returns = pd.concat((my_portfolio_returns,df_daily), axis = 1, join = "inner")
portfolio_returns.rename(columns = {0:"my_portfolio"}, inplace = True)
portfolio_returns
```

```python
#Drop NaNs
portfolio_returns.dropna(inplace=True)
portfolio_returns
```

```python
#annual STD
portfolio_std = pd.DataFrame(portfolio_returns.std()).rename(columns = {0:"std"})
portfolio_std
```

```python
#60 day Beta
rolling_covariance = portfolio_returns['my_portfolio'].rolling(window=60).cov(portfolio_returns['sp500'])
rolling_variance = portfolio_returns['sp500'].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of my portfolio')
```

```python
trading_days = portfolio_returns.count()
sharpe_ratios = portfolio_returns.mean()*trading_days/(portfolio_returns.std()*np.sqrt(trading_days))
sharpe_ratios.plot(kind = 'bar', title = "sharpe ratios result")
```
